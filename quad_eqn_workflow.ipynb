```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Literal
```

```python
class QuadState(TypedDict):
    a: int
    b: int
    c: int
    
    eqn: str
    discriminant: float
    res: str 
```

```python
def show_eqn(state: QuadState):
    eqn = f'{state['a']}x2{state['b']}x{state['c']}'
    return {'eqn': eqn}
def cal_discriminant(state: QuadState):
    d = state['b'] ** 2 - (4* state['a']*state['c'])
    return {'discriminant': d}
```

```python
def real_roots(state: QuadState):
    r1 = (-state['b'] + state['discriminant']**0.5)/(2 * state['a'])
    r2 = (-state['b'] - state['discriminant']**0.5)/(2 * state['a'])
    res = f'The roots are {r1} and {r2}'
    return {'res': res}

def repeated_roots(state: QuadState):
    r1 = (-state['b'])/(2 * state['a'])
    res = f'The roots are {r1} and {r1}'
    return {'res': res}

def no_roots(state: QuadState):
    res = f' No real roots'
    return {'res': res}

def check(state: QuadState) -> Literal['real_roots', 'repeated_roots', 'no_roots']:
    if state['discriminant'] > 0:
        return 'real_roots'
    elif state['discriminant'] == 0:
        return 'repeated_roots'
    else:
        return 'no_roots'
```

```python
graph = StateGraph(QuadState)

graph.add_node('show_eqn', show_eqn)
graph.add_node('cal_discriminant', cal_discriminant)

graph.add_node('real_roots', real_roots)
graph.add_node('repeated_roots', repeated_roots)
graph.add_node('no_roots', no_roots)

graph.add_edge(START, 'show_eqn')
graph.add_edge('show_eqn', 'cal_discriminant')

graph.add_conditional_edges('cal_discriminant', check)
graph.add_edge('real_roots', END)
graph.add_edge('repeated_roots', END)
graph.add_edge('no_roots', END)

wf = graph.compile()
```

```python
initial = {
    'a': 4,
    'b': -5,
    'c': -4
}
wf.invoke(initial)
```
