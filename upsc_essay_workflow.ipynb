```python
from langgraph.graph import StateGraph, START, END
from langchain_community.chat_models import ChatOllama
from typing import TypedDict, Annotated
from pydantic import BaseModel, Field
import operator
```

```python
model = ChatOllama(model="llama3")
```

```python
class EvaluationScheme(BaseModel):
    feedback : str = Field(description='Write a detailed feedback for the given essay')
    score: int = Field(description='Write the score out of 10 for the given essay', ge = 0, le= 10)
```

```python
struct_model = model.with_structured_output(EvaluationScheme)
```
