```python
from langgraph.graph import StateGraph, START, END
from langchain_community.chat_models import ChatOllama
from typing import TypedDict
```

```python
model = ChatOllama(model="llama3")
```

```python
class llm_state(TypedDict):
    ques : str
    ans : str
```

```python
def llm_qa(state: llm_state) -> llm_state:
    ques = state['ques']
    prompt = f'Answer the following question {ques}'
    ans = model.invoke(prompt)
    state['ans'] = ans.content
    return state
```

```python
graph = StateGraph(llm_state)
graph.add_node('llm_qa', llm_qa)
graph.add_edge(START, 'llm_qa')
graph.add_edge('llm_qa', END)
wf = graph.compile()   
```

```python
initial_state = {
    "ques": "What is the meaning of 'Priyal' as name?"
}
finalstate = wf.invoke(initial_state)
print(finalstate)
```
