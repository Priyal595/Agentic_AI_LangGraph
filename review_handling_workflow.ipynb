```python
from langgraph.graph import StateGraph, START, END
from langchain_community.chat_models import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from typing import TypedDict, Literal
from pydantic import BaseModel, Field
import json
```

```python
model = ChatOllama(model="llama3")
```

```python
class SentimentSchema(BaseModel):
    sentiment: Literal['positive', 'negative'] = Field(description='Sentiment of review')
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a sentiment classifier. Respond ONLY with valid JSON."),
    ("human", """
Classify the sentiment of the following text.

INPUT_TEXT

Return EXACTLY this JSON format (no extra text):
{ "sentiment": "positive" or "negative" }
""")
])
```

```python
struct_model = prompt | model
text = 'Classify the sentiment of the following text. - The software is too bad'
struct_model.invoke(text)
```
